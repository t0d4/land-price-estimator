```python
import os
import leafmap
from samgeo import SamGeo, show_image, download_file, overlay_images, tms_to_geotiff
import pyproj
from PIL import Image, TiffImagePlugin

sur = 0.005 #取得範囲
zoom = 16 #１７で保存するといい感じ（経度によるwidthの歪みがあるのでリサイズ必須）
siz = 512  #リサイズ

imagepathlist = []
images = []
for i in range(len(datalist)): #40をlen(datalist)
    m = leafmap.Map(center=[datalist[i,0],datalist[i,1] ], zoom=20, height="800px")
    m.add_basemap(f"SATELLITE{i}")
    bbox = [datalist[i,0]-sur,datalist[i,1]-sur ,datalist[i,0]+sur, datalist[i,1]+sur]
    #image = f"{filepath}/images/satellite{str(i).zfill(6)}.tif"
    #ローカル用
    image = f"images/satellite{str(i).zfill(6)}.tif"
    imagepathlist.append(image)
    tms_to_geotiff(output=image, bbox=bbox, zoom=zoom, source="Satellite", overwrite=True)
    img = Image.open(image)
    img = img.resize((siz,siz))
    img.save(image)
    img = np.array(img)
    images.append(img)
    imagepathlist.append(image)

images = np.stack(images)

imagepathlist = np.array(imagepathlist)
chika = datalist[:,2]
print(images.shape) #衛星画像
print(chika.shape)   #地価
np.save('images.npy', images)
np.save('chika.npy', chika)
```
